```python
import numpy as np
import sympy as sp
import random as rd
import tkinter as ttk
import time





main_root = ttk.Tk()
some_canv = ttk.Canvas(main_root,width = 500,height = 500,background = "green")
new_frame = ttk.Canvas(main_root,width = 400,height = 400,background = "black")
some_canv.create_window(250,250,window = new_frame,anchor = ttk.CENTER)
some_canv.grid(column = 0,row = 0,columnspan = 10,rowspan = 2)



def create_points(the_list,the_number):
	the_list.append((0,0))
	current = (0,0)
	for number in range(the_number):
		value = rd.random()
		if value > 0.93:
			the_x = current[0]
			the_y = current[1]
			new_x = -0.15*the_x+0.28*the_y
			new_y = 0.26*the_x+0.24*the_y+0.44
			the_list.append((new_x,new_y))
			current = (new_x,new_y)
		elif 0.86 < value <= 0.93:
			the_x = current[0]
			the_y = current[1]
			new_x = 0.2*the_x-0.26*the_y
			new_y = 0.23*the_x+0.22*the_y+1.6
			the_list.append((new_x,new_y))
			current = (new_x,new_y)
		elif 0.01 < value < 0.86:
			the_x = current[0]
			the_y = current[1]
			new_x = 0.85*the_x+0.04*the_y
			new_y = -0.04*the_x+0.85*the_y+1.6
			the_list.append((new_x,new_y))
			current = (new_x,new_y)
		elif value <= 0.01:
			the_x = current[0]
			the_y = current[1]
			new_x = 0
			new_y = 0.16*the_y
			the_list.append((new_x,new_y))
			current = (new_x,new_y)
	new_list = []
	for item in the_list:
		new_item = (item[0]*30+200,-item[1]*30+340)
		new_list.append(new_item)
	return new_list


def p_point(the_point):
	new_frame.create_oval(the_point[0],the_point[1],the_point[0]+0.1,the_point[1]+0.1,fill="white")

def paint(the_vals):
	for k in the_vals:
		time.sleep(0.000000001)
		p_point(k)
		new_frame.update()



def start():
	new_frame.delete("all")
	paint(create_points([],50000))

the_but = ttk.Button(main_root,text = "start",command = start)
the_but.grid(column = 4,row = 2,rowspan = 1)






main_root.mainloop()
```
